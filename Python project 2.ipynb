```python
# Descriptive analysis for non-categorical data - measures of central tendency
# Select non-categorical (numerical) columns
numerical_cols = hts.select_dtypes(include=['number'])

# Function to calculate measures of central tendency
def central_tendency_measures(hts):
    measures = pd.DataFrame({
        'Minimum': hts.min(),
        'Maximum': hts.max(),
        'Mean': hts.mean().round(2),
        'Median': hts.median(),
        'Mode': hts.mode().iloc[0],  # Mode may have multiple values, take the first one
        '25th Percentile': hts.apply(lambda x: np.percentile(x.dropna(), 25)),
        '50th Percentile': hts.apply(lambda x: np.percentile(x.dropna(), 50)),
        '75th Percentile': hts.apply(lambda x: np.percentile(x.dropna(), 75))
    })
    return measures

# Calculate and display the measures of central tendency for numerical columns
measures = central_tendency_measures(numerical_cols)
print(measures)
```

```python
# Descriptive analysis for non-categorical data - measures of dispersion

# Import relevant libraries
import scipy.stats as stats
from scipy.stats import skew, kurtosis

# Selecting non-categorical (numeric) columns
numeric_hts = hts.select_dtypes(include=[np.number])

# Function to calculate measures of dispersion
def calculate_dispersion(hts):
    dispersion_metrics = {}

    # Range
    dispersion_metrics['Range'] = hts.max() - hts.min()

    # Standard Deviation
    dispersion_metrics['Standard Deviation'] = hts.std().round(2)

    # Skewness
    dispersion_metrics['Skewness'] = hts.apply(lambda x: skew(x.dropna())).round(4)

    # Kurtosis
    dispersion_metrics['Kurtosis'] = hts.apply(lambda x: kurtosis(x.dropna(), fisher=False)).round(4)

    return pd.DataFrame(dispersion_metrics)

# Correlation matrix
correlation_matrix = numeric_hts.corr().round(4)

# Calculate measures of dispersion
dispersion_results = calculate_dispersion(numeric_hts)

# Display results
print("Dispersion Measures:")
print(dispersion_results)

print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Function to calculate Coefficient of Variation (CV)
def coefficient_of_variation(hts):
    return (hts.std() / hts.mean()) * 100
# Function to calculate 95% Confidence Interval
def confidence_interval(hts, confidence=0.95):
    ci_dict = {}
    for col in hts.columns:
        sample_data = hts[col].dropna()  # Remove NaN values
        mean = np.mean(sample_data)
        sem = stats.sem(sample_data)  # Standard Error of the Mean
        margin = sem * stats.t.ppf((1 + confidence) / 2., len(sample_data) - 1)
        ci_dict[col] = (round(mean - margin, 4), round( mean + margin, 4))
    return ci_dict

# Calculating Coefficient of Variation
cv_results = numeric_hts.apply(coefficient_of_variation).round(4)

# Calculating Confidence Interval
ci_results = confidence_interval(numeric_hts)


# Display Results
print("Coefficient of Variation (in %):")
print(cv_results)

print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci}")
```

```python
#Infrential Analysis - Test of Variance (Levene)

# Group data by Shipping Method
air_quantity = hts[hts['Shipping_Method'] == 'Air']['Quantity']
sea_quantity = hts[hts['Shipping_Method'] == 'Sea']['Quantity']
land_quantity = hts[hts['Shipping_Method'] == 'Land']['Quantity']

# Perform Levene's test to check for equality of variance across Shipping Methods
stat, p_value = stats.levene(air_quantity, sea_quantity, land_quantity)

# Print results
print(f"Levene’s test statistic: {stat}")
print(f"p-value: {p_value}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Variances are significantly different (reject null hypothesis).")
else:
    print("Variances are equal (fail to reject null hypothesis).")
```

```python
#z-test
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Count the number of Imports and Exports from the 'Import_Export' column in hts DataFrame
import_count = (hts['Import_Export'] == 'Import').sum()
export_count = (hts['Import_Export'] == 'Export').sum()

# Total number of transactions (Imports + Exports)
total_count = len(hts)

# Perform the z-test for proportions
counts = np.array([import_count, export_count])
nobs = np.array([total_count, total_count])

# Conduct z-test
z_stat, p_value = proportions_ztest(counts, nobs)

# Display results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in proportions.")
else:
    print("Fail to reject the null hypothesis: No significant difference in proportions.")
```

```python
#Test of Normality
#import relevant libraries
from scipy.stats import shapiro, kstest, anderson, jarque_bera

#Apply normality tests to Quantity, Value, and Weight
# 1. Shapiro-Wilk Test
print("Shapiro-Wilk Test:")
for column in ['Quantity', 'Value', 'Weight']:
    stat, p_value = shapiro(hts[column])
    print(f"{column}: statistic={round(stat, 4)}, p-value={round(p_value, 4)}")

# 2. Kolmogorov-Smirnov Test
# Assuming normal distribution parameters for ks test
print("\nKolmogorov-Smirnov Test:")
for column in ['Quantity', 'Value', 'Weight']:
    mean = np.mean(hts[column])
    std = np.std(hts[column], ddof=1)
    stat, p_value = kstest(hts[column], 'norm', args=(mean, std))
    print(f"{column}: statistic={round(stat, 4)}, p-value={round(p_value, 4)}")

# 3. Anderson-Darling Test
print("\nAnderson-Darling Test:")
for column in ['Quantity', 'Value', 'Weight']:
    result = anderson(hts[column], dist='norm')
    print(f"{column}: statistic={round(result.statistic, 4)}, critical values={result.critical_values}, significance levels={result.significance_level}")

# 4. Jarque-Bera Test
print("\nJarque-Bera Test:")
for column in ['Quantity', 'Value', 'Weight']:
    stat, p_value = jarque_bera(hts[column])
    print(f"{column}: statistic={round(stat, 4)}, p-value={round(p_value, 4)}")
```
